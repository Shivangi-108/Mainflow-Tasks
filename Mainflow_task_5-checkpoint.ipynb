```python
#1. How many people have heart disease and how many people doesn't have heart disease?

#getting the values
df.target.value_counts()
```

```python
#plotting bar chart

df.target.value_counts().plot(kind = 'bar', color=["orchid", "salmon"])
plt.title("heart disease values")
plt.xlabel("1 = Heart Disease, 0= No Heart Disease")
plt.ylabel("Amount")
```

```python
#pie chart
df.target.value_counts().plot(kind = 'pie', figsize=(8,6))
plt.legend(["Disease", "No Disease"]);
```

```python
# males and females
df.sex.value_counts()
```

```python

#piechart
df.sex.value_counts().plot(kind = 'pie', figsize=(8,6))
plt.title('Male female Ratio')
plt.legend(['Male', 'Female']);
```

```python
#2. People of which sex has the most heart disease?
pd.crosstab(df.target, df.sex)
```

```python
sns.countplot(x='target', data = df, hue = 'sex')
plt.title("heart disease frequency for sex")
plt.xlabel("0 = No heart disease, 1= heart disease");
```

```python
#3. People of which sex has which type of chest pain most'

df.cp.value_counts()
```

```python
#bar chart
df.cp.value_counts().plot(kind = 'bar', color = ['salmon', 'lightskyblue', 'springgreen','khaki'])
plt.title('Chest pain type vs count');
```

```python
#4. People with which chest pain are most prone to have heart disease?

pd.crosstab(df.cp, df.target)
```

```python
sns.countplot(x='cp', data =df, hue = 'target');
```

```python
#create a distribution plot with normal distribution curve
sns.displot(x='age', data = df, bins=30, kde = True);
```

```python
#create a distribution plot fro 'max heart rate
sns.displot(x='thalach', data = df, bins=30, kde = True, color='chocolate');
```

```python
#5. What is the correlation between resting blood pressure and the presence of heart disease
correlation = df[['trestbps', 'target']].corr()
```

```python
# Scatter plot to visualize the relationship
plt.figure(figsize=(10, 6))
sns.scatterplot(x='trestbps', y='target', data=df)
plt.title('Resting Blood Pressure vs Heart Disease Presence', fontsize=14)
plt.xlabel('Resting Blood Pressure (trestbps)', fontsize=12)
plt.ylabel('Presence of Heart Disease (target)', fontsize=12)
plt.grid(True)
plt.show()
```

```python
#6. What is the distribution of thalassemia types among individuals with and without heart disease?
thal_distribution = df.groupby(['thal', 'target']).size().unstack().fillna(0)
```

```python
# Bar plot to visualize the distribution
thal_distribution.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Distribution of Thalassemia Types Among Individuals With and Without Heart Disease', fontsize=14)
plt.xlabel('Thalassemia Type (thal)', fontsize=12)
plt.ylabel('Number of Individuals', fontsize=12)
plt.legend(['No Heart Disease', 'Heart Disease'], title='Target')
plt.grid(axis='y')
plt.show()
```

```python
#7. What is the pattern between ST depression and the presence of heart disease?
# Scatter plot to visualize the relationship between ST depression and heart disease presence
plt.figure(figsize=(10, 6))
sns.scatterplot(x='oldpeak', y='target', data=df, hue='target', palette='coolwarm', s=100)
plt.title('ST Depression vs Heart Disease Presence', fontsize=14)
plt.xlabel('ST Depression (oldpeak)', fontsize=12)
plt.ylabel('Presence of Heart Disease (target)', fontsize=12)
plt.grid(True)
plt.show()
```
